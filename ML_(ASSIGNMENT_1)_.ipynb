```python
# 16. Replace missing values and verify that no NaN values remain in the dataset.

# to replace missing values in symboling column as it is categorical we apply mode imputation
df['symboling'].fillna(df['symboling'].mode()[0],inplace=True)
# to replace missing values in normalized-losses column as it is shown in histogram that it is positively skewed we apply median imputation
df['normalized-losses'].fillna(df['normalized-losses'].median(),inplace=True)
# to replace missing values in make column as it is categorical we apply mode imputation
df['make'].fillna(df['make'].mode()[0],inplace=True)
# to replace missing values in fuel-type column as it is categorical we apply mode imputation
df['fuel-type'].fillna(df['fuel-type'].mode()[0],inplace=True)
# to replace missing values in aspiration column as it is categorical we apply mode imputation
df['aspiration'].fillna(df['aspiration'].mode()[0],inplace=True)
# to replace missing values in num-of-doors column as it is categorical we apply mode imputation
df['num-of-doors'].fillna(df['num-of-doors'].mode()[0],inplace=True)
# to replace missing values in body-style column as it is categorical we apply mode imputation
df['body-style'].fillna(df['body-style'].mode()[0],inplace=True)
# to replace missing values in drive-wheels column as it is categorical we apply mode imputation
df['drive-wheels'].fillna(df['drive-wheels'].mode()[0],inplace=True)
# to replace missing values in engine-location column as it is categorical we apply mode imputation
df['engine-location'].fillna(df['engine-location'].mode()[0],inplace=True)
# to replace missing values in wheel-base column as it is shown in histogram that it is positively skewed we apply median imputation
df['wheel-base'].fillna(df['wheel-base'].median(),inplace=True)
# to replace missing values in engine-type column as it is categorical we apply mode imputation
df['engine-type'].fillna(df['engine-type'].mode()[0],inplace=True)
# to replace missing values in num-of-cylinders column as it is categorical we apply mode imputation
df['num-of-cylinders'].fillna(df['num-of-cylinders'].mode()[0],inplace=True)
# to replace missing values in fuel-system column as it is categorical we apply mode imputation
df['fuel-system'].fillna(df['fuel-system'].mode()[0],inplace=True)
# applying median imputation on length , width ,curb-weight ,engine-size and height column as they are positively skwed
df['length'].fillna(df['length'].median(),inplace=True)
df['width'].fillna(df['width'].median(),inplace=True)
df['height'].fillna(df['height'].median(),inplace=True)
df['curb-weight'].fillna(df['curb-weight'].median(),inplace=True)
df['engine-size'].fillna(df['engine-size'].median(),inplace=True)
```

```python
skewness_A = df['bore'].skew()

print(f"Skewness of column 'A': {skewness_A}")
df['bore'].hist(bins=10, alpha=0.7, label='A')

plt.legend()
plt.show()
```

```python
# hence according to histogram bore is nearly symmetric we will apply mean imputation to column bore
df['bore'].fillna(df['bore'].mean(),inplace=True)
# hence according to histogram the column stroke is negatively skewed
df['stroke'].fillna(df['stroke'].median(),inplace=True)
# hence peak-rpm is nearly symetric we are applying mean imputation
df['peak-rpm'].fillna(df['peak-rpm'].mean(),inplace=True)
# hence column compression-ratio, horsepower, city-mpg, highway-mpg ,and price are positively skewed we are applying median imputation
columns_to_impute = ['compression-ratio', 'horsepower', 'city-mpg', 'highway-mpg','price']
df['compression-ratio'].fillna(df['compression-ratio'].median(),inplace=True)
df['horsepower'].fillna(df['horsepower'].median(),inplace=True)
df['city-mpg'].fillna(df['city-mpg'].median(),inplace=True)
df['highway-mpg'].fillna(df['highway-mpg'].median(),inplace=True)
df['price'].fillna(df['price'].median(),inplace=True)
```

**2.2 Encoding Categorical Features**





```python
# 17. List all categorical features that need to be encoded.

# columns: make,fuel-type,aspiration,num-of-doors,body-style,drive-wheels,engine-location,engine-type,num-of-cylinders,fuel-system
```

```python
# 18: Use One-Hot Encoding or Label Encoding for categorical columns. Explain why you chose your method.

# label encoding column (num-of-doors,num-of-cylinders)
""" comment : choose label encoding for these column as they are ordinal ( hierarchy , power dynamics between them ) """


from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['num-of-doors'] = encoder.fit_transform(df['num-of-doors'])
df['num-of-cylinders'] = encoder.fit_transform(df['num-of-cylinders'])
```

```python
#  : one hot encoding column  (make,fuel-type,aspiration,body-style,drive-wheels,engine-location,engine-type,fuel-system)

""" comment : choose one hot encoding for these column as they are nomial (all equal no hierarchy between them ) """
df = pd.get_dummies(df, columns=['make'], drop_first=True)
df = pd.get_dummies(df, columns=['fuel-type'], drop_first=True)
df = pd.get_dummies(df, columns=['aspiration'], drop_first=True)
df = pd.get_dummies(df, columns=['body-style'], drop_first=True)
df = pd.get_dummies(df, columns=['drive-wheels'], drop_first=True)
df = pd.get_dummies(df, columns=['engine-location'], drop_first=True)
df = pd.get_dummies(df, columns=['engine-type'], drop_first=True)
df = pd.get_dummies(df, columns=['fuel-system'], drop_first=True)
```

3.1 Applying Regression Models**

**LINEAR REGRESSION **`

```python
# 25. Train a Linear Regression model. Record the MSE and R2 Score.

import statsmodels.api as sm
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
# Print the parameters, i.e. the intercept and the slope of the regression line fitted
lr.params
```

```python
# Performing a summary operation lists out all the different parameters of the regression line fitted
print(lr.summary())
```

```python
# to calculate error term

y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)
```

```python
fig = plt.figure()
sn.distplot(res, bins = 15)
fig.suptitle('Error Terms', fontsize = 15)                  # Plot heading
plt.xlabel('y_train - y_train_pred', fontsize = 15)         # X-label
plt.show()
```

```python
# prediction on the test set
# Add a constant to X_test
X_test_sm = sm.add_constant(X_test)

# Predict the y values corresponding to X_test_sm
y_pred = lr.predict(X_test_sm)
```

```python
y_pred.head()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
#Returns the mean squared error; we'll take a square root
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
r_squared = r2_score(y_test, y_pred)
r_squared
```

**DECISION TREE REGRESSOR **

```python
# 26. Train a Decision Tree Regression model. Compare its performance with Linear Regression.

# choosing input and output

from sklearn.tree import DecisionTreeRegressor
# train the model
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
# to calculate r2 (that how well model learn the training data)
r_sq = model.score(X_train, y_train)
print(f"R2: {r_sq}")
```

```python
# mse of training data( less mean square error means model is working good )

from sklearn.metrics import mean_squared_error
from sklearn.tree import plot_tree

mean_squared_error(y_train, model.predict(X_train))
```

```python
# mse of testing data

mean_squared_error(y_test, model.predict(X_test)) #mean_squared_error(Y_true,Y_pred)
```

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV


# peforming hyper parameter tuining
param = {
    'criterion': ['squared_error', 'friedman_mse', 'poisson'],
    'max_depth': [None, 5, 20],
    'min_samples_split': [2, 5, 10],
    'splitter': ['best', 'random']
}
# Performing GridSearchCV
grid_search = GridSearchCV(model, cv=5, param_grid=param)
grid_search.fit(X_train, y_train)

# Best parameters and best score
print("Best Parameters:", grid_search.best_estimator_)
print("Best Score (R2):", grid_search.best_score_)

# Evaluating on test data
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Test Score (R2):", test_score)
```
